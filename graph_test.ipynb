```python
import networkx as nx
import random
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def ERgraph(N,p):
    # graph with N nodes
    nodes = range(0,N)
    G = nx.Graph()
    G.add_nodes_from(nodes)
    ## Create edges probablistically
    edges = [(i,j)
             for i in range(N)
             for j in range(N)
             if i < j and random.random() < p]
    G.add_edges_from(edges)
    ## Return the graph object
    return(G)
```

```python
G = ERgraph(100,0.05)
print("There are", G.number_of_nodes(), "nodes")
print("There are", G.number_of_edges(), "edges")
print("Draw the graph")
nx.draw(G)
```
